# 11월 5일 수업 _ sound

```python
# sinusoidal : 소리에서 나타나는 sin wave 같은 것을 지칭함
# phasor : 위의 sin function 과 같은 것을 만들어내는 것

## sin / cos 이해하기 ##
# degree:   0 ...180... 360 ...
# radians:  0 ...π... 2π ...
# sin 곡선 & cos 곡선 그래프 이해하기 (phasor)-> cos(90),cos(180),sin(0),sin(90)와 같이 대표적인 값들만 알면 됨!
# 0부터 100π까지 sin/cos곡선은 50번의 rythme이 나타난다! (2π마다 반복되기 때문)
# radian의 값은 cos(θ)라고 표기함. θ = 3/2π 이면 sin(θ)의 값은 -1. 이렇게 구할 수 있음
# 심화된 학습을 하고 싶다면 위키피디아에 '오일러 공식'을 검색! 

## 오일러 공식 ##
# e^(θi) = cos(θ) + sin(θ)*i = f(θ) 라고 정의.
# e와 허수 i의 값이 정해져있으므로 θ의 값에 따라 값을 출력해주는 함수라고 생각하면 됨!
# 복소수 : a + bi => 모든 수를 표현할 수 있음. 수의 범위에서 가장 큰 카테고리
# θ =    0, π/2, π, 3π/2, 2π
# f(θ) = 1,  i,   -1,  -i,    1  ....... 앞의 4개가 계속 반복되며 나올 것

## 복소수의 plot 방법 ##
# complex plain 을 사용함. x축이 a, y축이 b를 나타냄. 즉, 1은 (1,0), i는 (0,1)로 표기.
# complex plain에 위의 4개의 값을 그리고 이으면 반지름이 1인 원을 그림. 
# 원점에서 그 원에 그은 직선과 x축 사이의 각도는 θ가 됨. 
# (1,0) (0,1)과 같은 값들을 모두 벡터라고 할 수 있음. 

## projection ##
# a의 관점에서만 (즉, 실수의 관점에서만) 보겠다고 하면, b의 값, 즉 위아래로 왔다갔다하는 것처럼 보일 것 (sin(θ)의 그래프)
# b의 관점에서만 (즉, 허수의 관점에서만) 보겠다고 하면, a의 값, 즉 옆으로 왔다갔다하는 것처럼 보일 것 (cos(θ)의 그래프)
```

```python
# phasor를 생성할 때 각도의 개념과 더불어 시간의 개념을 같이 넣아줘야 함. 
# sin(θ)라고만 하면 소리의 실체를 만들어낼 수 없음. frequency를 통해 시간의 개념을 추가해줌
```

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.axes_grid1 import make_axes_locatable
import IPython.display as ipd
import numpy as np
%matplotlib notebook
from scipy.signal import lfilter
```

```python
# parameter setting
amp = 1          # range [0.0, 1.0]    -> 진폭을 정의
sr = 10000       # sampling rate, Hz    -> sampling rate은 음질의 해상도를 결정, 1초에 10000개의 숫자로 표현
dur = 0.5        # in seconds           -> 몇초동안 소리가 지속되는가
freq = 100.0     # sine frequency, Hz   -> 1초에 몇번 진동하는가 
```

```python
# generate time  => time을 먼저 만들어아 함
# t = 0.0001 0.0002 0.0003 ... 0.5000 의 time tick을 만드는 것 (0.5000는 duration, sampling rate는 10000)
t = np.arange(1, sr * dur+1)/sr # 위의 array를 간단하게 만들어내는 함수
t # e-04는 10^(-4)을 뜻함. 즉, 1.000e-04는 1.000*0.0001 = 0.0001을 뜻함.
```

```python
# generate phase
theta = t * 2*np.pi * freq
# 여기서 2*np.pi는 전체 하나의 harmonic을 말함. 여기다 freq를 곱해주면 이 harmonic을 freq만큼 반복시킨다는 뜻
theta # t 와 theta의 size(개수)는 당연히 같다.
```

```python
# generate signal by cosine-phasor
s = np.sin(theta)
s
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(t[0:1000], s[0:1000], '.')   # 2개의 입력값(time과 sin의 결과값)을 받음
ax.set_xlabel('time (s)') # x축에 label을 달아주는 것
ax.set_ylabel('real') # y축에 label을 달아주는 것
# 0.5까지는 결과값이 너무 많기 때문에 0.1까지의 결과값을 그래프로 보여줌
```

```python
# 0부터 2π까지 각도의 벡터를 만들기 위해서는 어떻게?
theta = np.arange(0,2*np.pi,0.1) # default increment가 1이므로 1,2,3,5,6까지 출력
theta # theta를 radian으로 정의함
```

```python
s = np.sin(theta)
s
```

```python
fig = plt.figure() # plt안에 있는 figure함수를 이용해 figure를 만들어놓음
ax = fig.add_subplot(221) # figure는 이 밑의 scatter plot전체를 말하는것 (4개 만들면 4개 출력)
ax.plot(theta, s, '.')    # theta가 x값, s가 y의 (결과)값 (즉, x의 값을 sin function에 통과시킨 것)
ax = fig.add_subplot(222) 
ax.plot(theta, s, '.')
ax = fig.add_subplot(223) 
ax.plot(theta, s, '.')
ax = fig.add_subplot(224) 
ax.plot(theta, s, '.')
# 태극문양처럼 되어 있는 것을 보아 sin함수의 형태이구나 라는 것을 알 수 있음 
# increment를 1로 하면 너무 sparse하다는 단점이 있음. np.arange()의 increment를 작게 해주면 더 빽빽하게 나올 것!
```

```python
fig = plt.figure()
ax = fig.add_subplot(111) 
ax.plot(theta, s, '-') # 여기서 '-'는 line으로 나타내는 것 (default), '.'로 하면 점으로 나타내는 것
ax.set_xlabel('theta in radians') 
ax.set_ylabel('value (sin(theta))')
# linear한 경우에 x축과 y축 모두 equidistance하고 (x와 y의 increment가 일정하다)
# nonlinear 한 경우에는 x축만 equidistance하고 y축은 아님. 밑의 예시
```

```python
# 만약 theta의 끝나는 지점이 2pi가 아니고 10pi면, 태극문양의 반복이 5번 반복될 것!
theta10 = np.arange(0,10*np.pi,0.1)
s10 = np.sin(theta10)
fig = plt.figure()
ax = fig.add_subplot(111) 
ax.plot(theta10, s10, '.')
ax.set_xlabel('theta in radians') 
ax.set_ylabel('value (sin(theta))')
```

```python
# 하.지.만 위의 그래프들은 시간이 빠져있는 형태이다!
# 시간을 추가하기 위해 먼저 time tick의 개수를 index로 먼저 만든다고 생각한다.
# 만약 1초라면 time tick의 개수는 sampling rate 와 일치한다. 
sr = 10000
dur = 0.5
t = np.arange(1, sr*1)
t = np.arange(1, sr * dur+1)/sr # dur을 생각한 time을 만듬
```

```python
# generate phase
freq = 100
theta = t * 2*np.pi * freq 
# time 을 연동시켜서 2pi를 곱하고, 그것을 몇 번 반복시킬건지 명시하기 위해 freq를 곱함
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(t[0:1000], s[0:1000], '.') # 이 그래프에 나오는 점들의 개수는 1000개이다. []안에 있는 수가 두개가 같아야 할 것.
ax.set_xlabel('time (s)')
ax.set_ylabel('real')
```

```python
# np.exp를 오일러수 e라고 생각! 즉, 밑의 식은 e^(theta*1j). 여기서 변수는 theta밖에 없음
c = np.exp(theta*1j)
c
# 복소수의 값들이 벡터로 쭈루룩 있는 것 
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')   # projection을 3차원으로
ax.plot(t[0:1000], c.real[0:1000], c.imag[0:1000], '.')
# c의 값들은 모두 복소수의 나오는데, a+bi의 형태로 나옴
# 여기서 a의 값은 c.real에 해당(실수부분), b의 값은 c.imag에 해당(허수부분)
ax.set_xlabel('time (s)')
ax.set_ylabel('real')
ax.set_zlabel('imag')
# 마찬가지로 총 1000개의 점이 찍힘
# 위에서 보면 real만 볼 수 있고, 앞에서 보면 imag만 볼 수 있음. 
# 즉, 보는 각도에 따라 sin 곡선(imag만 보는 경우)이 보이기도, cos 곡선(real만 보는 경우)이 보이기도 한다.
```
